```python
import requests
import json
```

```python
url = 'https://api.github.com'
user='glebsolyanik'
```

```python
r = requests.get(f'{url}/users/{user}/repos')
```

```python
with open('data.json', 'w') as f:
    json.dump(r.json(), f)
```
